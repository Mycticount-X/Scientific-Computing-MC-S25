# Least Square Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Data dummy
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 4, 5])

# 2. Ubah x menjadi matriks desain [x, 1]
X = np.vstack([x, np.ones(len(x))]).T  # X shape: (n, 2)


# 3. Hitung koefisien Least Squares
# Rumus: beta = (X^T X)^(-1) X^T y
beta = np.linalg.inv(X.T @ X) @ X.T @ y

# 4. Ambil slope (m) dan intercept (b)
m, b = beta
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# 5. Prediksi nilai y dari model
y_pred = m * x + b

# 6. Visualisasi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, y_pred, color='red', label='Regresi Linier')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Least Squares Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

üìå **Tujuan**
Mencari garis regresi terbaik y = mx + b yang meminimalkan jumlah kuadrat selisih antara nilai prediksi dan nilai asli.

---

üìÑ **Pseudocode**

Input: Array x\[n], Array y\[n]
Output: Slope m, Intercept b, Array y\_pred\[n]

1. Buat array X dengan dua kolom:

   * Kolom pertama: x
   * Kolom kedua: 1 (untuk bias/intercept)
     X = \[ \[x‚ÇÅ, 1],
     \[x‚ÇÇ, 1],
     ...
     \[x‚Çô, 1] ]

2. Hitung X\_transpose = Transpose dari X

3. Hitung XTX = X\_transpose \* X

4. Hitung Inverse\_XTX = Invers dari XTX

5. Hitung XTy = X\_transpose \* y

6. Hitung beta = Inverse\_XTX \* XTy

   * beta\[0] = m (slope)
   * beta\[1] = b (intercept)

7. Hitung y\_pred\[i] = m \* x\[i] + b untuk i = 1 to n

8. Return m, b, dan y\_pred



---

üß† **Penjelasan Konsep**

- Langkah 1 menyiapkan data ke dalam bentuk matrix agar bisa dihitung dengan persamaan matriks.
- Langkah 6 adalah inti dari Least Squares Method:
  \[
  \beta = (X^T X)^{-1} X^T y
  \]
- Hasilnya adalah `m` dan `b` yang digunakan untuk memprediksi y dari x.

---

‚úÖ **Catatan Tambahan**

- Pendekatan ini bisa diperluas untuk regresi multivariat.
- Jika data sangat besar atau X^T X tidak invertible, bisa gunakan metode lain (misalnya `np.linalg.lstsq`).
